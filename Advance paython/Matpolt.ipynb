```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create the dataset
data = {
    "student_id": [4, 5, 6, 7, 8],
    "location": ["Sydney", "Tokyo", "Berlin", "London", "Tokyo"],
    "age": [24, 24, 22, 23, 22],
    "sql_marks": [95, 99, 72, 97, 91],
    "excel_marks": [99, 95, 70, 90, 71],
    "python_marks": [55, 89, 99, 74, 79],
    "power_bi_marks": [82, 86, 79, 72, 80],
    "english_marks": [75, 82, 77, 85, 75]
}

df = pd.DataFrame(data)
df
```

```python
#Line Plot
plt.plot(df["age"],df["python_marks"],marker= "o",linestyle = ":",color = "blue")
plt.title("Paython Marks of Student")
plt.xlabel("Student Age")
plt.ylabel("Paython Marks")
plt.grid(True)
plt.show()
```

```python
#Bar Chart
plt.bar(df["location"],df["age"],color ="skyblue",width = 0.5)
plt.title("Location vs Age")
plt.xlabel("Location")
plt.ylabel("Student Age")
plt.show()
```

```python
#Bar Chart Horizental
plt.barh(df["location"],df["age"],color ="skyblue")
plt.title("Location vs Age")
plt.xlabel("Location")
plt.ylabel("Student Age")
plt.show()
```

```python
#multiple Line Plot
plt.plot(df["student_id"],df["python_marks"],label="python_marks")
plt.plot(df["student_id"],df["english_marks"],marker= "s",label="english_marks")
plt.title("Paython Marks vs English Marks")
plt.xlabel("Student Id")
plt.ylabel("Marks")
plt.legend()
plt.show()
```

```python
# Pie chart
df["location"].value_counts().plot.pie(autopct="%1.1f%%",startangle = 0)
plt.title("Student by Location")
plt.ylabel("") #Hide default y-label
plt.show()
```

```python
#Histogram ,count
plt.hist(df["age"], bins=5, color='lightgreen', edgecolor='black')
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Number of Students")
plt.show()
```

```python
plt.hist(df["age"], bins=2, color='skyblue', edgecolor='black')
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Number of Students")
plt.locator_params(axis='y', integer=True) # Set y-axis ticks to integers
plt.show()
```

```python
# Scatter Plot
plt.scatter(df["python_marks"], df["sql_marks"], color="purple",alpha = 0.5)
plt.title("Paython vs SQL Marks")
plt.xlabel("Paython Marks")
plt.ylabel("SQL Marks")
plt.show()
```

```python
# Box Plot
plt.boxplot([df["sql_marks"]], labels=["SQL"])
plt.title("Marks Distribution by Subject")
plt.ylabel("Marks")
plt.show()
```
