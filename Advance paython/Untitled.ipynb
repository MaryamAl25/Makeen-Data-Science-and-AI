```python
import requests
import pandas as pd

# Your API key
api_key = "fee588a85f2b5e4bf5c5a6e8d6662155"

# List of cities
cities = ["Muscat", "Salalah", "Sohar", "Nizwa"]

api_url = "http://api.weatherapi.com/v1/current.json"

weather_data = []

for city in cities:
    params = {
        "key": api_key,
        "q": city
    }
    response = requests.get(api_url, params=params)
    
    try:
        data = response.json()
    except:
        print(f"Failed to parse response for {city}")
        continue
    
    # Check if API returned an error
    if "error" in data:
        print(f"Error fetching {city}: {data['error']['message']}")
    else:
        info = {
            "city": data["location"]["name"],
            "region": data["location"]["region"],
            "country": data["location"]["country"],
            "local_time": data["location"]["localtime"],
            "temp_c": data["current"]["temp_c"],
            "temp_f": data["current"]["temp_f"],
            "condition": data["current"]["condition"]["text"],
            "wind_kph": data["current"]["wind_kph"],
            "humidity": data["current"]["humidity"],
            "feelslike_c": data["current"]["feelslike_c"]
        }
        weather_data.append(info)

# Save to CSV
if weather_data:
    df = pd.DataFrame(weather_data)
    df.to_csv("weather_multiple_cities.csv", index=False)
    print("Weather data saved to weather_multiple_cities.csv")
else:
    print("No data saved. Check your API key or city names.")
```

```python
import requests
import pandas as pd


API_KEY = "e2ff8e020abb574b3f2eb97c686b2ba6"  


city = input("Enter city name: ")

url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={API_KEY}&units=metric"


response = requests.get(url)


data = response.json()

if str(data.get("cod")) != "200":
    print("City not found or API key invalid.")
    print("Message from API:", data.get("message"))
else:

    city_name = data["name"]                      # city name returned from API
    temperature = data["main"]["temp"]           # current temperature
    feels_like = data["main"]["feels_like"]      # feels like temperature
    humidity = data["main"]["humidity"]          # humidity %
    description = data["weather"][0]["description"]  # weather description
    wind_speed = data["wind"]["speed"]           # wind speed


    print("\nWeather Report:")
    print("City:", city_name)
    print("Temperature:", temperature, "C")
    print("Feels Like:", feels_like, "C")
    print("Humidity:", humidity, "%")
    print("Weather:", description)
    print("Wind Speed:", wind_speed, "m/s")


    weather_row = {
        "City": city_name,
        "Temperature_C": temperature,
        "Feels_Like_C": feels_like,
        "Humidity_%": humidity,
        "Description": description,
        "Wind_Speed_m_s": wind_speed
    }

df = pd.DataFrame([weather_row])  # list of one dict => one-row DataFrame

df.to_excel("weather_report.xlsx", index=False)

print("\nData saved to 'weather_report.xlsx'")
```

```python
import requests
import csv

# API URL
api_url = "https://fakestoreapi.com/products"

# Fetch data from API
response = requests.get(api_url)
data = response.json()

# Extract relevant information
# Adjust this part based on the specific data you want from Fakestore API
product_data = [
    [item['id'], item['title'], item['price'], item['category'], item['description']]
    for item in data
]

# Save data to CSV
with open("fakestore_data.csv", mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    # Write headers
    writer.writerow(["ID", "Title", "Price", "Category", "Description"])
    # Write product data
    writer.writerows(product_data)

print("Data saved to fakestore_data.csv")
```
