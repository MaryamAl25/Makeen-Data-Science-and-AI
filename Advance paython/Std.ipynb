```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

scores = np.array([50, 60,60,70,70,70,80,80,90])
mean = np.mean(scores)
std = np.std(scores)
z_scores = np.array([(x - mean) / std for x in scores])

print("Mean:", mean)
print("Std:", std)
print("Z-scores:", z_scores)

# Plot histogram of z-scores
plt.hist(z_scores, bins=5, density=True, alpha=0.6, color='skyblue', edgecolor='black')

# Overlay the standard normal distribution curve
x = np.linspace(min(z_scores)-1, max(z_scores)+1, 1000)
plt.plot(x, norm.pdf(x, 0, 1), 'r', linewidth=2)  # mean=0, std=1 for z-scores

plt.title("Histogram of Z-scores with Normal Distribution")
plt.xlabel("Z-score")
plt.ylabel("Density")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#data
scores = np.array([50, 60,60,70,70,70,80,80,90])
mean = np.mean(scores)
std = np.std(scores)
z_scores = np.array([(x - mean) / std for x in scores])

print("Mean:", mean)
print("Std:", std)
print("Z-scores:", z_scores)

# Plot histogram
plt.hist(z_scores, bins=5, edgecolor='black', alpha=0.7)
plt.title("Histogram of Z-scores")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.xticks(z_scores)  # show exact z-scores on x-axis
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
