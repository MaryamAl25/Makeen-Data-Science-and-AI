```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample data
df = pd.DataFrame({
    'Employee_ID': range(1, 11),
    'Monthly_Expense': [295, 310, 320, 305, 315, 290, 300, 310, 325, 400]
})

# Apply log transformation
df['Log_Expense'] = np.log(df['Monthly_Expense'])

# Plot histograms
plt.figure(figsize=(10, 5))

# Before log transformation
plt.subplot(1, 2, 1)
plt.hist(df['Monthly_Expense'], bins=8, edgecolor='black')
plt.title('Before Log Transformation')
plt.xlabel('Monthly Expense')
plt.ylabel('Frequency')

# After log transformation
plt.subplot(1, 2, 1)
plt.hist(df['Log_Expense'], bins=8, edgecolor='black', color='orange')
plt.title('After Log Transformation')
plt.xlabel('Log(Monthly Expense)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample data
df = pd.DataFrame({
    'Employee_ID': range(1, 11),
    'Monthly_Expense': [295, 310, 320, 305, 315, 290, 300, 310, 325, 400]
})

# Apply log transformation
df['Log_Expense'] = np.log(df['Monthly_Expense'])

# Plot histograms
plt.figure(figsize=(10, 5))

# Common labels
x_label = 'Expense Value'
y_label = 'Frequency'

# Before log transformation
plt.subplot(1, 2, 1)
plt.hist(df['Monthly_Expense'], bins=8, edgecolor='black')
plt.title('Before Log Transformation')
plt.xlabel(x_label)
plt.ylabel(y_label)

# After log transformation
plt.subplot(1, 2, 2)
plt.hist(df['Log_Expense'], bins=8, edgecolor='black', color='orange')
plt.title('After Log Transformation')
plt.xlabel(x_label)
plt.ylabel(y_label)

plt.tight_layout()
plt.show()
```

```python
import scipy.stats as stats
import matplotlib.pyplot as plt

# Q-Q plot to check normality visually
stats.probplot(df['Log_Expense'], dist="norm", plot=plt)
plt.title("Q-Q Plot After Log Transformation")
plt.show()
```
