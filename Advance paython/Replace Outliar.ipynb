```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("data.xlsx")
df = pd.DataFrame(df)
df
```

```python
#IQR method
Q1 = df['Monthly_Expense'].quantile(0.25)
Q3 = df['Monthly_Expense'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

iqr_outliers = df[(df['Monthly_Expense'] < lower_bound) | (df['Monthly_Expense'] > upper_bound)]

print("IQR outliers:")
print(iqr_outliers)
```

```python
df['outliar_capped'] = np.where(df['Monthly_Expense'] > upper_bound, upper_bound,
                             np.where(df['Monthly_Expense'] < lower_bound, lower_bound, df['Monthly_Expense']))
df
```
