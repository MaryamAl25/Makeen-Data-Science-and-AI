```python
import numpy as np
```

```python
# Step 1: Create a 4x4 matrix with random numbers from 0 to 15
A = np.random.randint(0,16,(4,4))
print("Array 4x4 Matrix: \n", A)
```

```python
# Step 2: Print the second column
print("Second column: \n",A[:,1])
```

```python
# Step 3: Print the bottom two rows
print("Bottom two rows: \n",A[2:,:])
```

```python
# Step 4: Extract the submatrix #like: [[6,7],[10,11]]
submatrix = A[:2, 2:4]
print("\nSubmatrix (top-right 2x2):\n", submatrix)
```

```python
# Step 5: Horizontally split the matrix into two parts (left and right)
left, right = np.vsplit(A, [2]) #give the index rows by rows
print("\nleft:\n", left)
print("\nright:\n", right)
```

```python
# Step 6: Vertically split the matrix into two parts (upper and lower)
upper, lower = np.hsplit(A, 2) #give the index column by column
print("\nupper:\n", upper)
print("\nlower:\n", lower)
```

```python
# Step 7: Concatenate the two right halves back together vertically
print(np.concatenate([upper,lower],axis=1))
```
