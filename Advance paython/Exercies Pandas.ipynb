```python
import pandas as pd
import numpy as np
```

```python
#way2:downloading the data
df = pd.read_csv("data_science_student_marks.csv")
df
```

```python
#1 Inspect the Data
#Q1)Print the first 10 rows of the dataset.
print(df.head(10))
```

```python
#Q2)Print last 5 rows of the dataset.
print(df.tail(5))
```

```python
#Q3)Display column names, shape, and info.
print(df.columns,df.shape,df.info)
```

```python
#2Selecting Columns and Rows
#Q1)Display only the location and python_marks columns.
df.iloc[:,[1,5]]
```

```python
#Q2)Display the first 5 rows of sql_marks and excel_marks
df.iloc[:5,[3,4]]
```

```python
#Q3)Select the row for student_id 10.
df.loc[df["student_id"]==10]
```

```python
#3Filtering Rows
#Q1)Find all students who scored more than 90 in Python.
print(df[df["python_marks"] > 90])
```

```python
#Q2)Find students located in Tokyo with sql_marks 90.
print(df[(df["sql_marks"] == 90) & (df["location"] == "Tokyo")])
```

```python
#Q3)Find students whose age is 22 or 23.
print(df[(df["age"] == 22) | (df["age"] == 23)])
```

```python
#Q4)Find students who scored more than 80 in Python or more than 90 in Excel
print(df[(df["python_marks"] > 80) | (df["excel_marks"] > 90)])
```

```python
#4 Using .isin() , .between() , .str
#Q1)Select students located in London, Sydney, or Berlin.
df[df["location"].isin(["London", "Sydney", "Berlin"])]
```

```python
#Q2)Select students with age between 21 and 24.
df[df["age"].between(21, 24)]
```

```python
#5 Sorting
#Q1)Sort students by python_marks descending.
df.sort_values(by="python_marks", ascending=False)
```

```python
#Q2)Sort by location ascending and sql_marks descending.
df.sort_values(by=["location", "sql_marks"], ascending=[True, False])
```

```python
#Q3)Sort the DataFrame by age ascending, then english_marks descending
df.sort_values(by=["age", "english_marks"], ascending=[True, False])
```

```python
#6 Adding / Modifying Columns
#Q1)Create a column total_marks which is the sum of all marks columns.
df["total_marks"] = df[["python_marks", "sql_marks", "excel_marks", "power_bi_marks", "english_marks"]].sum(axis=1)
df
```

```python
#Q2)Create a column average_marks which is the mean of all marks columns.
df["average_marks"] = df[["python_marks", "sql_marks", "excel_marks", "power_bi_marks", "english_marks"]].mean(axis=1)
df
```

```python
#Q3)Create a column passed which is Yes if average_marks = 75 else No
df["passed"] = np.where(df["average_marks"] >= 75, "Yes", "No")
df
```

```python
#7 Dropping Columns and Rows
#Q1)Drop the column power_bi_marks
df.drop("power_bi_marks", axis=1, inplace=True)
df
```

```python
#Q2)Drop the row for student_id 496
df = df.drop(df[df["student_id"] == 496].index)
df
```

```python
#Q3)Drop the first 5 rows of the DataFrame
df = df.drop(df.index[:5])
df
```

```python
#8 Aggregation
#Q1)Calculate the average Python marks
df["python_marks"].mean()
```

```python
#Q2)Find the maximum Excel marks.
df["excel_marks"].max()
```

```python
#Q3)Calculate the average total marks by location.
df.groupby("location")["total_marks"].mean()
```

```python
#Q4)Count how many students scored more than 90 in SQL
(df["sql_marks"] > 90).sum()
```

```python
#9 Using .loc and .iloc
#Q1)Select rows 10 to 20 and columns 
df.loc[10:20, :]
```

```python
#Q2)Update english_marks of student_id to student_id 5 to 85.
df.loc[df["student_id"] == 5, "english_marks"] = 85
df
```

```python
#Q3)Select the 5th row, 3rd column using python_marks.iloc 
df.iloc[4, 2]
```

```python
#10 Challenge: Combining Skills
#Q1)Find the top 5 students with highest average marks
df.sort_values(by="average_marks", ascending=False).head(5)
```
