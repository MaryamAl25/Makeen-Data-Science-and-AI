```python
Creating and Accessing a Pandas Series

Task:
Create a Series that stores the monthly sales (in OMR) for 5 stores using this data:
stores = ['Muscat', 'Sohar', 'Nizwa', 'Salalah', 'Sur']
sales = [1200, 950, 1100, 1600, 800]

Display:

All store names (index).
All sales values.
The sales value for Salalah using label indexing.
The sales value of the last store using position indexing.
```

```python
sales = [1200, 950, 1100, 1600, 800]
stores = ['Muscat', 'Sohar', 'Nizwa', 'Salalah', 'Sur']
sales_series = pd.Series(sales, index=stores)
print("All store names (index):\n ",sales_series.index) #All store names (index).
print("-"*50)
print("All sales values: ",sales_series.values)#All sales values.
print("-"*50)
print("The sales value for Salalah: ", sales_series['Salalah']) #The sales value for Salalah using label indexing
print("-"*50)
print("The sales value of the last store: ",sales_series[-1]) #The sales value of the last store using position indexing.
#print("The sales value of the last store: ",sales_series[4])
```

```python
TaskB:
Using NumPy operations on the Series, compute:
The total sales across all stores.
The average sales.
The stores with sales above the average.
The difference of each store’s sales from the highest sale.
-------------------------------------------------------------------
```

Task 2:

```python
import numpy as np
```

```python
sales_series = pd.Series(sales, index=stores)
print("sales series: ",sales_series)
print("-"*50)
total_sales_series=np.sum(sales_series)
print("the total sales series: ", total_sales_series) #The total sales across all stores.
print("-"*50)
average__sales_series=np.mean(sales_series) #The average sales.
print("average sales series: ",average__sales_series)
print("-"*50)
above_avg=sales_series[sales_series>average__sales_series] #The stores with sales above the average.
print("sales above the average: ",above_avg) 
print("-"*50)
max_sales = np.max(sales_series)
sales_difference = max_sales - sales_series  #The difference of each store’s sales from the highest sale.
print("The difference of each store’s sales:\n",sales_difference)
```

```python
mySeries = pd.Series([10,20,30,40,50], name="Grades", index=["Ali", "Ahmed","Muna","Noor", "Nada"])
print(mySeries)

# 1 Series inside Series to filteration
boolean = mySeries > 40
print(mySeries[boolean])

# 2 Series inside Series to filteration
boolean = pd.Series([True, True, False, False, True], index=["Ali", "Ahmed","Muna","Noor", "Nada"])
print(mySeries[boolean])

# 3 List inside Series to filteration
booleanList = [True, True, False, False, True]
print(mySeries[booleanList])

# 4 Array inside Series to filteration
booleanSeries = pd.Series([True, True, False, False, True])
booleanValuesAsArray = booleanSeries.values
print(mySeries[booleanValuesAsArray])
```
