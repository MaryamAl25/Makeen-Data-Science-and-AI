```python
import numpy as np 
import pandas as pd 
```

```python
#Exercise 1: Mean, Median, and Mode 
gpas = [2.7, 3.0, 3.5, 3.8, 3.0, 3.2, 3.4, 3.0, 3.1, 3.9]

# Calculate values
mean_gpa = round(np.mean(gpas), 2)
median_gpa = round(np.median(gpas), 2)
mode_gpa = round(pd.Series(gpas).mode()[0], 2)

# Create DataFrame
results = pd.DataFrame({
    "Statistic": ["Mean", "Median", "Mode"],
    "Value": [mean_gpa, median_gpa, mode_gpa]
})

# Display DataFrame
print(results)
```

```python
#Exercise 2: Effect of Outliers 
salaries = [500, 520, 530, 540, 550, 9000] 

# Calculate statistics
mean_salary = round(np.mean(salaries), 2)
median_salary = round(np.median(salaries), 2)
mode_salary = round(pd.Series(salaries).mode()[0], 2)

# Create DataFrame
results = pd.DataFrame({
    "Statistic": ["Mean", "Median", "Mode"],
    "Value": [mean_salary, median_salary, mode_salary]
})

# Display DataFrame
print(results)
```

```python
#Exercise 3: Range, Variance, and Standard Deviation 
ages = [18, 19, 20, 20, 21, 22, 22, 23, 24, 30]

# Convert to DataFrame
df = pd.DataFrame({"Ages": ages})

# 1. Calculate statistics
age_range = round(df["Ages"].max() - df["Ages"].min(), 2)
variance = round(df["Ages"].var(), 2)      # sample variance (n-1)
std_dev = round(df["Ages"].std(), 2)       # sample standard deviation (n-1)

# 2. Create results DataFrame
results = pd.DataFrame({
    "Statistic": ["Range", "Variance", "Standard Deviation"],
    "Value": [age_range, variance, std_dev]
})

# Display 
print("\nResults:")
print(results)
```

```python
#Exercise 4: Comparing Two Groups
group_A = [65, 67, 68, 69, 70, 72, 74, 75] 
group_B = [60, 63, 67, 70, 75, 78, 80, 85] 

# Calculate statistics
mean_A = round(np.mean(group_A), 2)
std_A = round(np.std(group_A, ddof=1), 2)   # sample standard deviation

mean_B = round(np.mean(group_B), 2)
std_B = round(np.std(group_B, ddof=1), 2)

# Create a DataFrame
results = pd.DataFrame({
    "Group": ["A", "B"],
    "Mean": [mean_A, mean_B],
    "Standard Deviation": [std_A, std_B]
})

print(results)
```

```python
#Group A = steady performance
#Group B = mixed performance (less consistent)
```
