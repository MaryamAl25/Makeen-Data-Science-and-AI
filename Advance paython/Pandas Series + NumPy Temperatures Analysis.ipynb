```python
import pandas as pd
import numpy as np
```

```python
days = ["Day1","Day2","Day3","Day4","Day5","Day6","Day7"]
muscat_temps = [35,36,34,33,37,38,36]
sohar_temps  = [28,29,27,30,31,28,29]
nizwa_temps  = [40,42,39,41,43,40,41]
```

```python
#Q1: Create Series
muscat = pd.Series(muscat_temps, index=days, name="Muscat")
sohar  = pd.Series(sohar_temps,  index=days, name="Sohar")
nizwa  = pd.Series(nizwa_temps,  index=days, name="Nizwa")

print("Series for each city\n")
print(muscat, "\n")
print(sohar, "\n")
print(nizwa, "\n")
```

```python
#Q2: Average Temperature
avg_muscat = np.mean(muscat)
avg_sohar  = np.mean(sohar)
avg_nizwa  = np.mean(nizwa)

print("Average Temperatures (°C)")
print(f"Muscat: {avg_muscat:.1f}")
print(f"Sohar:  {avg_sohar:.1f}")
print(f"Nizwa:  {avg_nizwa:.1f}")
```

```python
cities_above_35 = [name for name, avg in (
    ("Muscat",avg_muscat), ("Sohar",avg_sohar), ("Nizwa",avg_nizwa)) if avg > 35]
print("Q3: Cities with average > 35°C:", ", ".join(cities_above_35), "\n")
```

```python
cities = {
    "Muscat": avg_muscat,
    "Sohar": avg_sohar,
    "Nizwa": avg_nizwa
}

cities_above_35 = []
for name, avg in cities.items():
    if avg > 35:
        cities_above_35.append(name)

print("Cities with average > 35°C:", ", ".join(cities_above_35), "\n")
```

```python
#Q4: Hottest day per city
print("Hottest day per city")
print(f"Muscat: {muscat.idxmax()} ({muscat.max()}°C)")
print(f"Sohar : {sohar.idxmax()} ({sohar.max()}°C)")
print(f"Nizwa : {nizwa.idxmax()} ({nizwa.max()}°C)\n")
```

```python
#Q5: Boolean Series above average
muscat_above_avg = muscat > avg_muscat
sohar_above_avg  = sohar  > avg_sohar
nizwa_above_avg  = nizwa  > avg_nizwa

print("Boolean Series (True = Above city average)\n")
print("Muscat:", muscat_above_avg, "\n")
print("Sohar:", sohar_above_avg, "\n")
print("Nizwa:", nizwa_above_avg, "\n")
```

```python
#Q6: Count of above-average days
print("Count of above-average days")
print(f"Muscat: {muscat_above_avg.sum()}")
print(f"Sohar : {sohar_above_avg.sum()}")
print(f"Nizwa : {nizwa_above_avg.sum()}\n")
```

```python
#Q7: Count days within 35°C–40°C
muscat_in_range = ((muscat >= 35) & (muscat <= 40)).sum()
sohar_in_range  = ((sohar  >= 35) & (sohar  <= 40)).sum()
nizwa_in_range  = ((nizwa  >= 35) & (nizwa  <= 40)).sum()

print("Days with temperatures between 35°C and 40°C (inclusive)")
print(f"Muscat: {muscat_in_range}")
print(f"Sohar : {sohar_in_range}")
print(f"Nizwa : {nizwa_in_range}\n")
```

```python
#Q8: Compare two cities
muscat_hotter_than_sohar = muscat > sohar
print("Days when Muscat was hotter than Sohar:\n", muscat_hotter_than_sohar,"\n")
print("Count:", muscat_hotter_than_sohar.sum(), "days\n")
```

```python
#Q9: Extreme hot days (> 1.1 × average)
muscat_extreme = muscat > (1.1 * avg_muscat)
sohar_extreme  = sohar  > (1.1 * avg_sohar)
nizwa_extreme  = nizwa  > (1.1 * avg_nizwa)

print("Extreme hot days (> 10% above average)\n")
print("Muscat:\n", muscat_extreme, "\nCount:", muscat_extreme.sum(), "\n")
print("Sohar:\n", sohar_extreme, "\nCount:", sohar_extreme.sum(), "\n")
print("Nizwa:\n", nizwa_extreme, "\nCount:", nizwa_extreme.sum(), "\n")
```

```python
#Q10: Coolest day overall
df = pd.DataFrame({"Muscat": muscat, "Sohar": sohar, "Nizwa": nizwa})
overall_min = df.min().min()
city_min = df.min().idxmin()
day_min = df[city_min].idxmin()

print("Coolest day overall")
print(overall_min,"°C",city_min,day_min)
```

```python
#Q11: Difference to average & largest positive difference 
muscat_diff = muscat - avg_muscat
sohar_diff  = sohar  - avg_sohar
nizwa_diff  = nizwa  - avg_nizwa

print("Q11: Difference from average and largest positive difference day\n")
print("Muscat differences:", muscat_diff, "\nLargest positive diff:", muscat_diff.idxmax(), f"({muscat_diff.max():.1f}°C)\n")
print("Sohar differences:", sohar_diff, "\nLargest positive diff:", sohar_diff.idxmax(), f"({sohar_diff.max():.1f}°C)\n")
print("Nizwa differences:", nizwa_diff, "\nLargest positive diff:", nizwa_diff.idxmax(), f"({nizwa_diff.max():.1f}°C)\n")
```
