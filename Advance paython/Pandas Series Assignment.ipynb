Pandas Series Assignment: 
Boolean Indexing and Filtering

```python
import pandas as pd
```

```python
mySeries = pd.Series([35, 28, 40, 32,30], index=['Muscat', 'Sohar', 'Nizwa', 'Salalah','Sur'],name="Daily Temperature")
print(mySeries)
```

```python
# Task 1: Filter using a condition
# Filter the Series for temperatures greater than 32.
```

```python
temperatures=mySeries[(mySeries > 32)]
temperatures
```

```python
# Task 2: Filter using a custom boolean Series
# Use a custom boolean Series to filter the data.
```

```python
booleanSeries = pd.Series([True, False, True, False, True],index=['Muscat', 'Sohar', 'Nizwa', 'Salalah','Sur'])
mySeries[booleanSeries] # here will check the index postion
```

```python
# Task 3: Filter using a list of booleans
# Use a boolean list to filter the Series
```

```python
boolean_list = [True, False, True, False, True]
filtered_series = mySeries[boolean_list]
print(filtered_series)
```

```python
#Task 4: Filter using a boolean array
#Use a boolean array to filter the Series
```

```python
import numpy as np
boolean_array = np.array([True, False, True, False, True])
filtered_series2 = mySeries[boolean_array]
print(filtered_series2)
```

```python
#Extra Challenge:
 # Filter the Series for temperatures between 30 and 36.
```

```python
abovea=mySeries[(mySeries >= 30) & (mySeries <= 36)]
abovea
```
