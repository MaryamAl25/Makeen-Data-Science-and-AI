```python
import pandas as pd
import numpy as np
```

```python
from sklearn.impute import SimpleImputer
```

```python
# Create sample data
data = {
    'student_id': [1, 2, 3, 4, 5, 6, 7, 8],
    'name': ['Ali', 'Sara', 'Hassan', 'Mona', 'Khalid', 'Fatma', 'Omar', 'Layla'],
    'age': [22, np.nan, 24, 23, np.nan, 25, 21, np.nan],
    'python_marks': [85, 90, 78, 92, 88, 95, 80, 87],
    'excel_marks': [75, 82, 70, 89, 85, 90, 72, 88],
    'location': ['Muscat', 'Sohar', 'Nizwa', 'Salalah', 'Muscat', 'Sohar', 'Nizwa', 'Muscat']
}
```

```python
# Create DataFrame
df = pd.DataFrame(data)
print(df)
```

```python
imputer = SimpleImputer(strategy='mean')
```

```python
df['column_with_missing'] = imputer.fit_transform(df[['age']]).ravel()
df
```

```python
df.isna().sum() 
```

```python
df1 = pd.DataFrame(data)
print(df1)
```

```python
#fill in with value
df1["age"].fillna(24) 
df
```

```python
#backward fill
df1["age"] = df1["age"].bfill()
df1
```

```python
#forward fill
df1["age"] = df1["age"].ffill()
df1
```
