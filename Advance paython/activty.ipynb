```python
import pandas as pd
import numpy as np
```

```python
# Create sample data
data = {
    'student_id': [1, 2, 3, 4, 5, 6, 7, 8],
    'name': ['Ali', 'Sara', 'Hassan', 'Mona', 'Khalid', 'Fatma', 'Omar', 'Layla'],
    'age': [22, np.nan, 24, 23, np.nan, 25, 21, np.nan],
    'python_marks': [85, 90, np.nan, 92, 88, np.nan, 80, 87],
    'excel_marks': [75, 82, 70, np.nan, 85, 90, np.nan, 88],
    'location': ['Muscat', 'Sohar', np.nan, 'Salalah', 'Muscat', 'Sohar', 'Nizwa', np.nan]
}
```

```python
# Create DataFrame
df = pd.DataFrame(data)
print(df)
```

```python
#Q1. Detect Missing Values
# Display which cells in the DataFrame contain missing values and count how many are missing in each column.(Hint: Use isnull() and sum().)
df.isna().sum() 
```

```python
#Q2. Drop Missing Values
#Remove all rows where any column has a missing value.How many rows remain after this operation?(Hint: Use dropna().)
df.dropna(inplace=False)
```

```python
#Q3. Fill with a Static Value
#Replace all missing values in the location column with 'Unknown'.(Hint: Use fillna('Unknown').)
df["location"].fillna("Unknown", inplace=True)
df
```

```python
#Q4. Fill with Mean
#Fill all missing values in the age column with the average (mean) age of the available students.(Hint: Use fillna(df['age'].mean()).)
df["age"] = df["age"].fillna(df["age"].mean())
df
```

```python
#Q5. Fill with Median
#Fill missing values in the excel_marks column using the median of that column.(Hint: Use df['excel_marks'].fillna(df['excel_marks'].median()).)
df["excel_marks"] = df["excel_marks"].fillna(df["excel_marks"].median())
df
```

```python
#Q6. Forward Fill (ffill)
#Use forward fill to fill the missing values in the python_marks column.(Hint: Use fillna(method='ffill').)
df["python_marks"] = df["python_marks"].ffill()
df
```

```python
#Q7. Fill missing values in the age column using backward fill.(Hint: Use fillna(method='bfill'))
df["age"] = df["age"].bfill()
df
```
