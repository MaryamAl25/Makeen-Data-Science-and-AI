```python
import pymysql
import pandas as pd
from datetime import datetime
```

```python
# Sample data to store in MySQL
# -------------------------------
data = pd.DataFrame({
    "customer_id": ["C001", "C002", "C003", "C004", "C005"],
    "product": ["Laptop", "Mouse", "Keyboard", "Monitor", "Printer"],
    "quantity": [1, 2, 1, 3, 1],
    "timestamp": [
        "2025-11-16 10:15:00",
        "2025-11-16 11:20:00",
        "2025-11-16 12:30:00",
        "2025-11-16 13:45:00",
        "2025-11-16 14:50:00"
    ],
    "store": ["Muscat Store", "Seeb Store", "Salalah Store", "Nizwa Store", "Sohar Store"]
})

# -------------------------------
# Connect to MySQL database
# -------------------------------
central_db_connection = pymysql.connect(
    host="localhost",       # replace with your host
    user="root",            # replace with your MySQL username
    password="aaaa",    # replace with your MySQL password
    database="central_db"   # replace with your database name
)

cursor = central_db_connection.cursor()

# -------------------------------
# Create table if it doesn't exist
# -------------------------------
create_table_query = """
CREATE TABLE IF NOT EXISTS consolidated_transactions (
    id INT AUTO_INCREMENT PRIMARY KEY,
    customer_id VARCHAR(255),
    product VARCHAR(255),
    quantity INT,
    timestamp DATETIME,
    store VARCHAR(255)
);
"""
cursor.execute(create_table_query)

# -------------------------------
# Insert data into MySQL table
# -------------------------------
insert_query = """
INSERT INTO consolidated_transactions (customer_id, product, quantity, timestamp, store)
VALUES (%s, %s, %s, %s, %s);
"""

for _, row in data.iterrows():
    cursor.execute(insert_query, tuple(row))

central_db_connection.commit()
central_db_connection.close()

print("Data successfully stored in the central MySQL database.")
```

```python
import pandas as pd
import pymysql

# -------------------------------
# Read Excel data
# -------------------------------
df = pd.read_excel('oman_news_images.xlsx')

# Replace NaN with None for MySQL
df = df.where(pd.notnull(df), None)

# -------------------------------
# Connect to MySQL
# -------------------------------
central_db_connection = pymysql.connect(
    host="localhost",
    user="root",
    password="aaaa",
    database="central_db"
)

cursor = central_db_connection.cursor()

# -------------------------------
# Create table
# -------------------------------
create_table_query = """
CREATE TABLE IF NOT EXISTS news (
    id INT AUTO_INCREMENT PRIMARY KEY,
    Date TIMESTAMP DEFAULT CURRENT_TIMESTAMP,
    Title VARCHAR(255),
    Summary TEXT,
    Article_URL TEXT,
    Image_URL TEXT,
    Image_Alt TEXT
);
"""
cursor.execute(create_table_query)

# -------------------------------
# Insert data into table
# -------------------------------
insert_query = """
INSERT INTO news (Title, Summary, Article_URL, Image_URL, Image_Alt)
VALUES (%s, %s, %s, %s, %s);
"""

for _, row in df.iterrows():
    cursor.execute(insert_query, (
        row['Title'],
        row['Summary'],
        row['Article_URL'],
        row['Image_URL'],
        row['Image_Alt']
    ))

central_db_connection.commit()
central_db_connection.close()

print("Data successfully stored in the central MySQL database.")
```
